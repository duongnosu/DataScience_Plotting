Each row in the assignment datafile corresponds to a single observation.

The following variables are provided to you:

* **id** : station identification code
* **date** : date in YYYY-MM-DD format (e.g. 2012-01-24 = January 24, 2012)
* **element** : indicator of element type
    * TMAX : Maximum temperature (tenths of degrees C)
    * TMIN : Minimum temperature (tenths of degrees C)
* **value** : data value for element (tenths of degrees C)

```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd
import numpy as np
import datetime as dt
%matplotlib notebook
def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinSize_d{}.csv'.format(binsize))

    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()


#leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')

```

```python

def clean_df():
    df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
    #df.keys()
    df.sort_values(['Date', 'ID'],inplace=True) #sort df by Date and ID 

    df = df[df['Date'].str.endswith("-02-29") == False] #Drop entries with 02-29

    #df['Date'] = pd.to_datetime(df['Date']) #Convert Date to pandas DateTime function
    df['Data_Value'] = df['Data_Value'].apply(lambda x:float(x)/10) #convert from tenth to actual temp
    return df

#Seperate Day, Month, Year 
#df['*Day'] = df['Date'].dt.day
#df['Month'] = df['Date'].dt.month
#df['Year'] = df['Date'].dt.year




```

```python
def get_df_05to14():
    df = clean_df()
    df['Date'] = pd.to_datetime(df['Date'])
    df = df[(df['Date'].dt.year >= 2005) & (df['Date'].dt.year <= 2014)]
    return df
#get_df_05to14()
```

```python
def get_df_2015():
    #df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
    #df.keys()
    #df.sort_values(['Date', 'ID'],inplace=True) #sort df by Date and ID 

    #df = df[df['Date'].str.endswith("-02-29") == False] #Drop entries with 02-29

    #df['Date'] = pd.to_datetime(df['Date']) #Convert Date to pandas DateTime function
    #df['Data_Value'] = df['Data_Value'].apply(lambda x:float(x)/10) #convert from tenth to actual temp
    
    df = clean_df()
    df['Date'] = pd.to_datetime(df['Date'])
    df = df[df['Date'].dt.year == 2015]
    return df
#get_df_2015()
```

```python
def get_TMAX_05to14():
    df = get_df_05to14()
    
    df = df[df['Element'] == 'TMAX']
    df['Month&Day'] = df['Date'].apply(lambda x: x.strftime('%m-%d'))  
    df = df.groupby(['Month&Day'])[['Data_Value']].max()
    return df
#get_TMAX_05to14()
```

```python
def get_TMIN_05to14():
    df = get_df_05to14()
    df = df[df['Element'] == 'TMIN']
    df['Month&Day'] = df['Date'].apply(lambda x: x.strftime('%m-%d'))
    df = df.groupby(['Month&Day'])[['Data_Value']].min()
    return df
```

```python

def get_TMAX_2015():
    df = get_df_2015()
    df = df[df['Element'] == 'TMAX']
    df['Month&Day'] = df['Date'].apply(lambda x: x.strftime('%m-%d'))
    #df = df.groupby(['Month&Day'])[['Data_Value']].max()
    df = df.groupby(['Month&Day'])[['Data_Value']].max()
    return df
        
```

```python
def get_TMIN_2015():
    df = get_df_2015()
    df = df[df['Element'] == 'TMIN']
    
    df['Date'] = pd.to_datetime(df['Date'])
   
    df['Month&Day'] = df['Date'].apply(lambda x: x.strftime('%m-%d')) 
    df = df.groupby(['Month&Day'])[['Data_Value']].min()
    
    #This also work if you want to try
    #df = df.groupby('Month&Day').aggregate({'Data_Value':np.min})
    
    #df = df.rename(index=str, columns={'Data_Value':'TempC'}, inplace = True)
  
    return df
```

```python
max_0514 = get_TMAX_05to14()
min_0514 = get_TMIN_05to14()



max_2015 = get_TMAX_2015()
min_2015 = get_TMIN_2015()

#Rename for readable
max_0514 = max_0514.rename(index=str, columns={'Data_Value':'TempC'})
min_0514 = min_0514.rename(index=str, columns={'Data_Value':'TempC'})

max_2015 = max_2015.rename(index=str, columns={'Data_Value':'TempC_15'})
min_2015 = min_2015.rename(index=str, columns={'Data_Value':'TempC_15'})

#y_max = max_0514['Data_Value']
#y_min = min_0514['Data_Value']

join_max = pd.concat([max_0514,max_2015], axis=1)
join_min = pd.concat([min_0514,min_2015], axis=1)

#join_min.head()
```

```python
outlier_max = np.where(join_max['TempC_15'] > join_max['TempC'])
outlier_min = np.where(join_min['TempC_15'] < join_min['TempC'])
#outlier_min

```

```python
def plot_fig():
    fig = plt.figure(1,figsize=(9, 6))
    #plt.figure(figsize=(9,6))
    max_temp = plt.scatter(outlier_max, join_max.iloc[outlier_max]
                           ['TempC_15'], alpha=1, color='#660708', marker='^',
                          label = 'Temperature in 2015 higher than period 2005-2014', zorder = 2)


    min_temp = plt.scatter(outlier_min, join_min.iloc[outlier_min]
                           ['TempC_15'], alpha=0.8, color='#577590', marker='v',
                          label = 'Temperature in 2015 lower than period 2005-2014', zorder = 2)

    x = np.arange(0, 365)
    x_axis = np.arange(0,365, 30.5)
    plt.xticks(x_axis, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], rotation=20, alpha=0.8)

    plt.yticks(alpha=0.8)

    
    plt.plot(x,max_0514.values, lw = 1.4, color='#ef476f',
            label = 'Maximum temperature for period 2005-2014', zorder = 1)
    plt.plot(x,min_0514.values, lw = 1.4, color='#43aa8b',
            label = 'Minimum temperature for period 2005-2014', zorder = 1)
    plt.fill_between(x, min_0514['TempC'], max_0514['TempC'], color='#bee3db', alpha = 0.5 , label = 'Temperature record range for period 2005-2014')


    plt.xlabel("Month of the year", fontsize= 10)
    plt.ylabel("Temperature (Celcius)", fontsize= 10)
    plt.title("Record highest & lowest temperature for each day of the year (2005-2014) \n Recorded by stations near Ann Arbor, Michigan", fontsize = 10)

    plt.legend(fontsize = 'x-small' )
    plt.gca().spines['right'].set_visible(False)
    plt.gca().spines['top'].set_visible(False)
    plt.savefig('assignment2.png')
    plt.show()
```

```python
plot_fig()
```
